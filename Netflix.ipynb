Bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Dataset

```python
filename = 'netflix_titles.csv'
```

```python
df = pd.read_csv(filename)
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

Limpeza

```python
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df.dropna(subset=['date_added', 'duration', 'rating'], inplace=True)
```

data

```python
df['date_added'] = df['date_added'].str.strip()
df['date_added'] = pd.to_datetime(df['date_added'])
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
```

```python
df.head()
```

```python
df.info()
```

Grafico

```python
sns.set(style="whitegrid")
```

```python
print('Proporção de filmes e séries na plataforma Netflix')
type_counts = df['type'].value_counts()
print(type_counts)
```

```python
plt.figure(figsize=(6, 6))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=140, colors=['#b20710', '#221f1f'])
plt.title('Proporção de Filmes e Séries na Plataforma Netflix')
plt.ylabel('')
plt.show()
```

```python
print('Top 10 paises com mais produções na Netflix')
top10_countries = df[df['country'] != 'Unknown']['country'].value_counts().head(10)
print(top10_countries)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=top10_countries.values, y=top10_countries.index, palette='viridis')
plt.xlabel('Número de Produções')
plt.ylabel('País')
plt.title('Top 10 Países com Mais Produções na Netflix')
plt.show()
```

```python
print('Conteudo adicionado por ano')
content_year = df['year_added'].value_counts().sort_index()
print(content_year)
```

```python
plt.figure(figsize=(12, 6))
content_year.plot(kind='line', marker = 'o', color='#b20710')
plt.xlabel('Ano')
plt.ylabel('Número de Conteúdos Adicionados')
plt.title('Conteúdos Adicionados por Ano')
plt.xticks(content_year.index)
plt.grid(True)
plt.show()
```
